```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
df1
```

```python
pd.merge(df1,df2,on='key')
```

```python
df3 = pd.DataFrame({'lkey':['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df3
```

df4

```python
df4
```

```python
pd.merge(df4,df3,left_on='rkey',right_on='lkey')
```

```python
print('上記mergeのデフォルトではinner joinが行われるため共通項目しか残らない..c,dは消える')
```

```python
pd.merge(df4,df3,left_on='rkey',right_on='lkey',how='outer') #outer join
```
